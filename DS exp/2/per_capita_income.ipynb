```python
# Loading dataset
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv("/content/canada_per_capita_income.csv")
print("Intial data : \n",df.head())
print("\nShape od data : ",df.shape)
```

```python
# Features and target
X = df[['year']]
y = df['per capita income (US$)']

# Train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Visualization
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```

```python
# Predictions
predicted_income = model.predict([[2025]])
print("Predicted Per Capita Income for 2025:", predicted_income[0])

print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_[0])
```
